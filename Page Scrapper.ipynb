```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
link = "https://quotes.toscrape.com/"
res = requests.get(link)
print(res.text)
```

```python
html = res.text
fd = open("main.html", "w", encoding="utf-8")
fd.write(html)
fd.close()
```

```python
soup = BeautifulSoup(res.text, "html.parser")
print(soup.prettify())
```

```python
quotes = soup.find_all("span", class_ = "text")
```

```python
for quote in quotes:
    print(quote.text)
```

```python
data = {"Quote": [], "Author": [], "URL": [], "tags": [], "Author DOB": [], "Author Birth-place": [], "Author Description": []}
for i in tqdm(range(1, 11), desc="Scraping pages", position=0, leave=False, colour="green"):
    url = f"https://quotes.toscrape.com/page/{i}/"
    res = requests.get(link).text
    soup = BeautifulSoup(res, "lxml")

    details = soup.find_all(class_ = "quote")
    for detail in details:
        quote = detail.find("span", class_ = "text").text[1:-1]
        author = detail.find("small").text
        author_id = detail.find("a").get("href")
        tags = list()
        for tag in detail.find_all("a", class_ = "tag"):
            tags.append(tag.text)
    
        tags = ",".join(tags)
        data["Quote"].append(quote)
        data["Author"].append(author)
        data["tags"].append(tags)

        link = "https://quotes.toscrape.com" + author_id
        about = requests.get(link).text
        soup2 = BeautifulSoup(about, "lxml")
        data["URL"].append(link)
     
        data["Author DOB"].append(soup2.find("span", class_ = "author-born-date").text)
        data["Author Birth-place"].append(soup2.find("span", class_ = "author-born-location").text)
        data["Author Description"].append(soup2.find(class_ = "author-description").text[9:-10])
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv("Quotes.csv", index = False)
```
