```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import re
from urllib.parse import unquote
```

```python
name = input("Enter a name: ").replace(" ", "+")
search = 'https://www.google.com/search?q=' + str(name) +" "+ "+wikipedia"
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
res = requests.get(search, headers=headers).text
soup = BeautifulSoup(res, "lxml")
print(soup.prettify())
```

```python
for link in soup.find_all("a"):
    href = link.get("href")
    if href and "en.wikipedia.org/wiki/" in href:
        url = href[7:]
        break
```

```python
url = url.split('&')[0]
url
```

```python
wiki = requests.get(url).text
```

```python
soup = BeautifulSoup(wiki, "lxml")
print(soup.prettify())
```

```python
heading = soup.find("h1").text
```

```python
corpus = ""
for p in soup.find_all("p"):
    corpus += p.text + "\n"
corpus = corpus.strip()
```

```python
print(corpus)
```

```python
corpus = re.sub(r"\[\d+\]", "", corpus)
```
