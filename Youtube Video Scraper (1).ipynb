```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
import pandas as pd
import numpy as np
```

```python
link = "https://www.youtube.com/@GeeksforGeeksVideos/videos"
```

```python
from selenium import webdriver
import chromedriver_binary
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager as CDM
driverpath = CDM().install()
```

```python
options = Options()
options.add_argument("--headless=new")           # Prevents opening multiple chrome tabs
options.add_argument("--disable-gpu")            # Prevents use of gpu for rendering
options.add_argument("--disable-extensions")     # Speed boost
options.add_argument("--disable-infobars")       # Removes i am a bot tag
options.add_argument("--mute-audio")             # Disables all output audio while scraping
options.add_argument("--window-size=1920,1080")  # Normal size chrome tab
```

```python
browser = webdriver.Chrome(service = Service(driverpath))
browser.get(link)
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
data = []

for sp in soup.find_all('ytd-rich-item-renderer'):
    
    title      = sp.find('a', class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    video_link = sp.find('a', class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')

    try:
        views  = sp.find_all('span', class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[0].text
    except:
        views  = np.nan
    try:
        date_time  = sp.find_all('span', class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[1].text
    except:
        date_time  = np.nan
        
    try:
        thumbnail_link = sp.find('img').get('src').split('?')[0]
    except:
        thumbnail_link = np.nan

        
    data.append([title, views, date_time, video_link, thumbnail_link])
```

```python
df = pd.DataFrame(data, columns = ['title','views','date_time','video_link','thumbnail_link'])
df
```

```python
df.to_csv("youtube_cover_page.csv", index=False)
```

```python
browser = webdriver.Chrome(service = Service(driverpath))
browser.get('https://www.youtube.com')
time.sleep(2)

data = []

for link in tqdm(df['video_link']):
    
    link = 'https://www.youtube.com' + link
    browser.get(link)
    
    time.sleep(5)
    
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    try:
        title = soup.find_all('yt-formatted-string', class_ = 'style-scope ytd-video-primary-info-renderer')[0].text
    except:
        title = np.nan
    
    try:
        view = soup.find('span' , class_ = 'view-count style-scope ytd-video-view-count-renderer').text
    except:
        view = np.nan
        
    try:
        date_time = soup.find_all('yt-formatted-string', class_ = 'style-scope ytd-video-primary-info-renderer')[1].text
    except:
        date_time = np.nan
    
    try:
        like = soup.find_all('div', class_ = 'yt-spec-button-shape-next__button-text-content')[4].text
    except:
        like = np.nan
      
    try:
        description = ""
        for sp in soup.find_all('yt-attributed-string', class_ = 'style-scope ytd-text-inline-expander'):
            description += sp.text.replace("\n\n", "")
    except:
        description = np.nan
        
    data.append([title , date_time, like, view, link, description])
```
